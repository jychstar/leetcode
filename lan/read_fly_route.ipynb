```python
file = "GX2029 ZGNNZSLG.txt"
info = dict(
    airline = file[0:2],
    flight_number = file[2:6],
    take_off = file[7:11],
    landing = file[11:15])
print(info)
results = []
flight_route = f"{info['airline']}-{info['take_off']}-{info['landing']}"
results.append([info['take_off'], info['landing'], flight_route,0,info['take_off'],0,0,0])

with open(file,'r') as fh:
    main_flag = False # used to indicate targeted text block
    waypoint_flag = False # used to indicate whether waypoint is found
    end_flag = False # used to indicate the last airport
    all_lines = fh.readlines()
    for line in all_lines:
        words = line.strip().split()
        if len(words)<=1:
            continue
        if words[0] == "MAIN" and words[2] == f"{info['take_off']}-{info['landing']}":
            print("MAIN PLAN is found and matched")
            main_flag = True
            counter = 0
        elif main_flag and words[0][0].isalpha() and words[1].isdigit():
            waypoint = words[0]
            elevation = words[1]
            #print(words) 
            #print(waypoint,elevation)
            waypoint_flag = True
        elif waypoint_flag and words[0][0] in "NS" and words[1][0] in "EW":
            # change sign if S and W
            lat = words[0][1:] if words[0][0] == "N" else "-" + words[0][1:]
            lon = words[1][1:] if words[1][0] == "E" else "-" + words[1][1:]
            counter +=1
            results.append([info['take_off'], info['landing'],flight_route,counter,waypoint,lat,lon,elevation])
            waypoint_flag = False 
            if waypoint == info['landing']:
                break # mission complete
        else:
            continue
print(counter)
```

```python
import pandas as pd
df = pd.DataFrame.from_records(results, 
    columns=['take_off', 'landing', 'flight_route','counter','waypoint','lat','lon','elevation'])
df
```
