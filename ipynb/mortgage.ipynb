## 2011, mortgage 500k

```python
# created in 2014.8.30
# modified in 2015.2.12
# adapted to python in 2019.9.25

current_remain = 500e3
monthly_pay = 6000 # monthly pay
total_month = 120
monthly_rate = 0.0063
print("first month interest:", current_remain * monthly_rate)
all_interest = []
all_remain = []
valid_month = []
for i in range(1,total_month):
    current_interest = current_remain * monthly_rate
    current_remain = current_remain + current_interest - monthly_pay
    if current_remain < 0:
        break
    all_interest.append(current_interest)
    all_remain.append(current_remain)
    valid_month.append(i)
print("total interest: ", sum(all_interest))
print("last month:{}, current remain:{} ".format(i,current_remain))
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi = 100)
plt.title("2011, mortgage 500k,monthly pay = 6k")
plt.plot(valid_month,all_remain,'o')
plt.xlabel("month")
plt.ylabel("Remain (CNY)")
plt.grid()
plt.show()
```
