## 2011, mortgage 500k

```python
# created in 2014.8.30
# modified in 2015.2.12
# adapted to python in 2019.9.25

current_remain = 500
monthly_pay = 6 # monthly pay
total_month = 120
monthly_rate = 0.0063
print("first month interest:", current_remain * monthly_rate)
all_interest = []
all_remain = []
valid_month = []
for i in range(1,total_month):
    current_interest = current_remain * monthly_rate
    current_remain = current_remain + current_interest - monthly_pay
    if current_remain < 0:
        break
    all_interest.append(current_interest)
    all_remain.append(current_remain)
    valid_month.append(i)
print("total interest: ", sum(all_interest))
print("last month:{}, current remain:{} ".format(i,current_remain))
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi = 100)
plt.title("2011, mortgage 500k,monthly pay = 6k")
plt.plot(valid_month,all_remain,'o')
plt.xlabel("month")
plt.ylabel("Remain (k CNY)")
plt.grid()
plt.show()
```

## 2011, personal loan
1. beginning from 2011.1, loan is 120 k, monthly rate is 1%. to 2015.1, the cumulative amount is 193 k

```python
120 * (1.01**48)
```

2. from 2015.1 to 2016.1, monthly payment is 3 k, the remaining amount is 180 k

```python
current_remain = 193
monthly_pay = 3 # monthly pay
total_month = 12
monthly_rate = 0.01
print("first month interest:", current_remain * monthly_rate)
all_interest = []
all_remain = []
valid_month = []
for i in range(1,total_month):
    current_interest = current_remain * monthly_rate
    current_remain = current_remain + current_interest - monthly_pay
    if current_remain < 0:
        break
    all_interest.append(current_interest)
    all_remain.append(current_remain)
    valid_month.append(i)
print("total interest: ", sum(all_interest))
print("last month:{}, current remain:{} ".format(i,current_remain))
```

3. from 2016.1 to 2019.9, continue to accumulate, total amount is 280 K

```python
180 * (1.01**44)
```

4. in 2019, 20 k has been paid, so the remaining amount is 260 k. 
---
- If monthly payment is 10 k, it will be paid off in 30 months. 
- If monthly payment is  8 k, it will be paid off in 40 months.
- If monthly payment is  6 k, it will pay off in 60 months.

```python
current_remain = 260
monthly_pay = 10 # monthly pay
total_month = 12*3
monthly_rate = 0.01
print("first month interest:", current_remain * monthly_rate)
all_interest = []
all_remain = []
valid_month = []
for i in range(1,total_month):
    current_interest = current_remain * monthly_rate
    current_remain = current_remain + current_interest - monthly_pay
    if current_remain < 0:
        break
    all_interest.append(current_interest)
    all_remain.append(current_remain)
    valid_month.append(i)
print("total interest: ", sum(all_interest))
print("last month:{}, current remain:{} ".format(i,current_remain))
```
